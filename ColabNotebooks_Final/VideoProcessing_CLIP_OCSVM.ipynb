```python
import cv2
from scenedetect import open_video, SceneManager, ContentDetector
import os

def segment_and_save_shots(video_path, output_dir, threshold=30.0):
    """Detects scenes in a video and saves frames from each detected scene."""
    os.makedirs(output_dir, exist_ok=True)
    video = open_video(video_path)
    scene_manager = SceneManager()
    scene_manager.add_detector(ContentDetector(threshold=threshold))
    scene_manager.detect_scenes(video)

    # Get the list of detected scenes
    scene_list = scene_manager.get_scene_list()

    # Capture frames from the original video
    cap = cv2.VideoCapture(video_path)

    for i, (start, end) in enumerate(scene_list):
        # Set the video capture to the start frame of the scene
        cap.set(cv2.CAP_PROP_POS_FRAMES, start.get_frames())
        ret, frame = cap.read()

        if ret:
            # Save the frame as an image file
            cv2.imwrite(os.path.join(output_dir, f"frame_{i}.jpg"), frame)

    # Release resources
    cap.release()

    return scene_list
```

```python
def process_multiple_movies_and_trailers(movie_paths, trailer_paths, output_base_dir, threshold=30.0):
    """Processes multiple movies and trailers and saves their shot frames."""

    # Combine movies and trailers for processing
    for movie_path in movie_paths:
        # Extract movie name without extension for directory naming
        movie_name = os.path.splitext(os.path.basename(movie_path))[0]
        output_dir = os.path.join(output_base_dir, movie_name, 'shots')

        print(f"Processing movie: {movie_name}...")
        scenes = segment_and_save_shots(movie_path, output_dir, threshold)

        # Print out the detected scenes
        for i, (start, end) in enumerate(scenes):
            print(f"Movie Scene {i}: Start Frame {start.get_frames()}, End Frame {end.get_frames()}")

    for trailer_path in trailer_paths:
        # Extract trailer name without extension for directory naming
        trailer_name = os.path.splitext(os.path.basename(trailer_path))[0]
        output_dir = os.path.join(output_base_dir, trailer_name, 'shots')

        print(f"Processing trailer: {trailer_name}...")
        scenes = segment_and_save_shots(trailer_path, output_dir, threshold)

        # Print out the detected scenes
        for i, (start, end) in enumerate(scenes):
            print(f"Trailer Scene {i}: Start Frame {start.get_frames()}, End Frame {end.get_frames()}")
```

```python
import os
import torch
import clip
from PIL import Image
import numpy as np

# Load the CLIP model and device setup
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)

def extract_clip_features_for_all_frames(shot_dir):
    """
    Extracts CLIP features for all frames in a single shot directory.
    Returns the average feature embedding for the entire shot directory.
    """
    features = []
    for frame_filename in sorted(os.listdir(shot_dir)):
        frame_path = os.path.join(shot_dir, frame_filename)
        if os.path.isfile(frame_path):
            image = preprocess(Image.open(frame_path)).unsqueeze(0).to(device)

            with torch.no_grad():
                frame_features = model.encode_image(image)
                features.append(frame_features.cpu().numpy())

    if features:
        # Average features for all frames in the shot to get one feature vector for the shot
        shot_embedding = np.mean(features, axis=0)
        return shot_embedding
    else:
        print(f"No frames found in {shot_dir}")
        return None

def extract_features_for_movie_or_trailer(shot_dir, output_path):
    """
    Extracts features for a movie or trailer by processing frames in a single directory,
    saving the averaged feature embeddings as a .npy file.
    """
    os.makedirs(os.path.dirname(output_path), exist_ok=True)

    shot_embedding = extract_clip_features_for_all_frames(shot_dir)

    if shot_embedding is not None:
        np.save(output_path, shot_embedding)
        print(f"Saved features for {shot_dir} to {output_path}")
    else:
        print(f"No features extracted for {shot_dir}, skipping.")

# Base directories
movies_base_dir = "/content/output_frames"
trailers_base_dir = "/content/output_frames"
output_movies_dir = "/content/features/movies"
output_trailers_dir = "/content/features/trailers"

# List of movie and trailer names
movie_names = ["BehindTheWalls", "HighWay", "NightVisit", "Stucco", "SushiNoh", "TheBottom", "TheChair", "UntilDeath"]
trailer_names = ["BehindTheWallsT", "HighWayT", "NightVisitT", "StuccoT", "SushiNohT", "TheBottomT", "TheChairT", "UntilDeathT"]
for movie_name, trailer_name in zip(movie_names, trailer_names):
    movie_shot_dir = os.path.join(movies_base_dir, movie_name, "shots")
    trailer_shot_dir = os.path.join(trailers_base_dir, trailer_name, "shots")

    movie_feature_path = os.path.join(output_movies_dir, f"{movie_name}.npy")
    trailer_feature_path = os.path.join(output_trailers_dir, f"{trailer_name}.npy")

    print(f"Processing movie: {movie_name}")
    extract_features_for_movie_or_trailer(movie_shot_dir, movie_feature_path)

    print(f"Processing trailer: {trailer_name}")
    extract_features_for_movie_or_trailer(trailer_shot_dir, trailer_feature_path)
```

```python
import os
import numpy as np
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler
import joblib  # for saving the model

def load_features(features_dir, name):
    feature_path = os.path.join(features_dir, f"{name}.npy")
    return np.load(feature_path)

# Example directories
trailer_features_dir = "/content/features/trailers"
movie_features_dir = "/content/features/movies"

# Load trailer features
trailer_features = []
trailer_names = ["BehindTheWallsT", "HighWayT", "NightVisitT", "StuccoT", "SushiNohT", "TheBottomT", "TheChairT", "UntilDeathT"]

for trailer_name in trailer_names:
    features = load_features(trailer_features_dir, trailer_name)
    trailer_features.append(features)

# Combine all trailer features into one array
trailer_features = np.vstack(trailer_features)

# Load movie features
def load_movie_features(features_dir):
    movie_features = []
    for movie_name in os.listdir(features_dir):
        if movie_name.endswith('.npy'):
            features = np.load(os.path.join(features_dir, movie_name))
            movie_features.append(features)
    return np.vstack(movie_features)

movie_features = load_movie_features(movie_features_dir)

# Combine movie and trailer features
combined_features = np.vstack([movie_features, trailer_features])

labels = np.array([1] * movie_features.shape[0] + [-1] * trailer_features.shape[0])

# Scale the combined features
scaler = StandardScaler()
combined_features_scaled = scaler.fit_transform(combined_features)

# Train One-Class SVM on the combined features
model = OneClassSVM(kernel='rbf', gamma='auto')
model.fit(combined_features_scaled[labels == -1])

# Save the trained model and scaler
joblib.dump(model, "v3CLIP2Model.pkl")
joblib.dump(scaler, "v3CLIP2Scaler.pkl")

print("Model and scaler saved successfully.")
```

```python
import os
import cv2
import pandas as pd
import numpy as np
import torch
import clip
from PIL import Image
import joblib

# Load the CLIP model and device setup
device = "cuda" if torch.cuda.is_available() else "cpu"
clip_model, preprocess = clip.load("ViT-B/32", device=device)

# Load the SVM model and scaler
svm_model = joblib.load("/content/video3CLIPSVCModel.pkl")
scaler = joblib.load("/content/video3CLIPSVCScaler.pkl")

def extract_features_for_segment(video_path, start_time, end_time):
    """
    Extracts CLIP features for frames within the specified segment of the video.
    """
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    start_frame = int(start_time * fps)
    end_frame = int(end_time * fps)
    features = []

    cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

    for frame_num in range(start_frame, end_frame + 1):
        ret, frame = cap.read()
        if not ret:
            break

        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        pil_image = Image.fromarray(frame_rgb)
        image_input = preprocess(pil_image).unsqueeze(0).to(device)

        with torch.no_grad():
            frame_features = clip_model.encode_image(image_input).cpu().numpy()
            features.append(frame_features)

    cap.release()

    if features:
        segment_embedding = np.mean(features, axis=0)
        return segment_embedding
    else:
        print(f"No features extracted for the segment from {start_time} to {end_time}.")
        return None

def predict_trailer_worthiness(features, threshold=-0.5):
    """
    Predicts trailer-worthiness using the decision function scores and a threshold.
    Returns both the binary prediction and the score.
    """
    # Get decision function score
    score = svm_model.decision_function(features)[0]

    # Convert score to prediction based on threshold
    prediction = 1 if score > threshold else 0

    return prediction, score

# Define the paths
csv_file = "/content/inTheDark.csv"
video_path = "/content/inTheDarkV.mp4"

# Read timestamps from the CSV
timestamps = pd.read_csv(csv_file)


DECISION_THRESHOLD = -1.07

results = []

# Main prediction loop
for index, row in timestamps.iterrows():
    start_time = row['Start Time (s)']
    end_time = row['End Time (s)']

    features = extract_features_for_segment(video_path, start_time, end_time)

    if features is not None:
        # Scale the features
        features_scaled = scaler.transform(features.reshape(1, -1))

        # Get prediction and score
        predicted_label, score = predict_trailer_worthiness(features_scaled, DECISION_THRESHOLD)

        # Append results
        results.append({
            'Start Time (s)': start_time,
            'End Time (s)': end_time,
            'Prediction': predicted_label,
            'Score': score
        })

        # Print the prediction result with score
        status = "trailer-worthy" if predicted_label == 1 else "not trailer-worthy"
        print(f"Timestamp {index}: The segment is {status} (score: {score:.3f})")
    else:
        print(f"Failed to extract features for timestamp {index}.")

# Save the prediction results to a CSV file
result_df = pd.DataFrame(results)
result_csv_path = "/content/prediction_results.csv"
result_df.to_csv(result_csv_path, index=False)

print(f"Prediction results saved to {result_csv_path}.")
```

```python
from moviepy.editor import VideoFileClip, concatenate_videoclips

# Load the movie file
movie = VideoFileClip("/content/inTheDarkV.mp4")

# Define the segments with start and end times (excluding "No use scene")
segments = [
    {"start": (5, 15), "end": (5, 20)},  # In Trailer
    {"start": (6, 45), "end": (6, 50)},  # Trailer Worthy, Not in Trailer
    {"start": (4, 35), "end": (4, 40)},  # In Trailer
    {"start": (5, 30), "end": (5, 35)},  # Key Moment, Not in trailer
    {"start": (2, 50), "end": (2, 55)},  # Key Audio
    {"start": (5, 50), "end": (5, 55)},  # Frame in Trailer
    {"start": (4, 15), "end": (4, 20)},  # in Trailer, Trailer Worthy
    {"start": (3, 0), "end": (3, 5)},    # Key Audio
    {"start": (0, 0), "end": (0, 5)},    # Movie Title
]

# Convert minutes and seconds to total seconds, then extract clips
clips = []
for segment in segments:
    start_time = segment["start"][0] * 60 + segment["start"][1]
    end_time = segment["end"][0] * 60 + segment["end"][1]
    # Extract the video clip and set its corresponding audio
    clip = movie.subclip(start_time, end_time).set_audio(movie.audio.subclip(start_time, end_time))
    clips.append(clip)

# Concatenate all clips in the correct order
final_clip = concatenate_videoclips(clips)

# Export the final video
final_clip.write_videofile("trailer.mp4", codec="libx264", audio_codec="aac")
```
